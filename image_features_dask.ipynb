```python
from dask.array.image import imread
import cv2
from dask.diagnostics import ProgressBar
import numpy as np
import os
import pandas as pd
import gc
gc.collect()
```

```python
def auto_canny_sub(image, sigma=0.33):
    # compute the median of the single channel pixel intensities
    v = np.median(image)
 
    # apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv2.Canny(image, lower, upper)
 
    # return the edged image
    return edged

def auto_canny(x):
    gray = cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (3, 3), 0)

    # apply Canny edge detection using a wide threshold, tight
    # threshold, and automatically determined threshold
    auto = auto_canny_sub(blurred)
    result = (float(np.sum(auto)) / (x.shape[0]*x.shape[1]))
    return result

def get_features(x):
    x = x[0, :, :]
    a_canny = auto_canny(x)
    #b_score = get_blurrness_score(x)
    #avg_color = get_average_color(x)
    #w_h = get_width_height(x)
    return np.array([[[[a_canny]]]])
```

```python
#IMAGE_DIR = '/home/ubuntu/.kaggle/competitions/avito-demand-prediction/test_jpg/data/competition_files/test_jpg/'
IMAGE_DIR = '/Users/Rashmi/Downloads/sample_trainimages1/'
im = imread( IMAGE_DIR + '*.jpg', imread=cv2.imread)  
print("Read images.",im)
result = im.map_blocks(get_features, dtype=im.dtype, chunks=(3500,1,1,1))

s = None
with ProgressBar():
    s = result.compute()
```
